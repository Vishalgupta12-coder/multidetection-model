```python
#import modules
import cv2
import tkinter as tk
from tkinter import *
#haarcascade files
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
body_cascade = cv2.CascadeClassifier('haarcascade_fullbody.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')
font = cv2.FONT_HERSHEY_SIMPLEX

#functions to detect different parameters
def face_detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.putText(frame,'face',(x,y),font,1,(0,0,255),2)
    return frame
def eye_detect(gray, frame):
    eyes = eye_cascade.detectMultiScale(gray, 1.1, 3)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(frame, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
        cv2.putText(frame,'eye',(ex,ey),font,1,(0,0,255),2)
    return frame
def body_detect(gray, frame):
    body = body_cascade.detectMultiScale(gray, 1.4, 4)
    for (bx, by, bw, bh) in body:
        cv2.rectangle(frame, (bx, by), (bx+bw, by+bh), (0, 255, 0), 2)
        cv2.putText(frame,'body',(bx,by),font,1,(0,0,255),2)
    return frame
def smile_detect(gray, frame):
    smile = smile_cascade.detectMultiScale(gray, 1.8, 5)
    for (sx, sy, sw, sh) in smile:
        cv2.rectangle(frame, (sx, sy), (sx+sw, sy+sh), (0, 255, 0), 2)
        cv2.putText(frame,'smile',(sx,sy),font,1,(0,0,255),2)
    return frame
#functions to open the camera and capture the video
def video():
    video_capture = cv2.VideoCapture(0)
    while True:
        _, frame = video_capture.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        canvas = face_detect(gray, frame)
        cv2.imshow('FACE DETECTION', canvas)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    video_capture.release()
    cv2.destroyAllWindows()
def video1():
    video_capture = cv2.VideoCapture(0)
    while True:
        _, frame = video_capture.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        canvas = eye_detect(gray, frame)
        cv2.imshow('EYES DETECTION', canvas)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    video_capture.release()
    cv2.destroyAllWindows()
def video2():
    video_capture = cv2.VideoCapture(0)
    while True:
        _, frame = video_capture.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        canvas = body_detect(gray, frame)
        cv2.imshow('BODY DETECTION',canvas)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    video_capture.release()
    cv2.destroyAllWindows()
def video3():
    video_capture = cv2.VideoCapture(0)
    while True:
        _, frame = video_capture.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        canvas = smile_detect(gray, frame)
        cv2.imshow('SMILE DETECTION', canvas)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    video_capture.release()
    cv2.destroyAllWindows()
    
window=tk.Tk()
window.geometry('600x400+750+100')
window.maxsize(1000,1200)
window['bg']='#D6E8C7'
window.title('MFDS')
window.iconbitmap('icon.ico')
# Add image file
bg = PhotoImage(file = "icon_r.png")
  
# Show image using label
label1 = Label( window, image = bg,width=180,height=180)
label1.place(x = 320, y = 90)
#buttons

l1=tk.Label(window,text='Multifactor Detection System',font=('Times New Roman',20,'bold'),bg='#D6E8C7')
l1.pack()

l_msg=tk.Label(window,text='(Press q to close the video)',font=('Times New Roman',15,'bold'),bg='#D6E8C7')
l_msg.pack()

b1=tk.Button(window,text='Detect Face',bg='lightblue',font=('bold'),width=15,command=video,bd=7)
b1.pack()
b1.place(x=30,y=80)

b2=tk.Button(window,text='Detect Eyes',bg='lightblue',font=('bold'),width=15,command=video1,bd=7)
b2.pack()
b2.place(x=30,y=145)

b3=tk.Button(window,text='Detect Smile',bg='lightblue',font=('bold'),width=15,command=video3,bd=7)
b3.pack()
b3.place(x=30,y=215)

b4=tk.Button(window,text='Detect Body',bg='lightblue',font=('bold'),width=15,command=video2,bd=7)
b4.pack()
b4.place(x=30,y=285)

b5=tk.Button(window,text='Exit',bg='#d0b48e',font=('bold'),command=window.destroy,width=10,bd=7)
b5.pack()
b5.place(x=350,y=285)

window.mainloop()
```
